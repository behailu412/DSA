# Student Records Sorting Program

This Python program manages and sorts student records using different sorting algorithms based on user selection. The program presents a menu-driven interface that allows users to sort student data by ID (using Bubble Sort), by name (using Insertion Sort), or by CGPA in descending order (using Selection Sort).

## Features

1. **Data Structure**: 
   - Stores student records as a list of dictionaries, with each dictionary containing:
     - `id`: Student ID (string)
     - `name`: Student name (string)
     - `dept`: Department (string)
     - `cgpa`: Cumulative GPA (float)

2. **Menu System**:
   - Presents a continuous loop menu with these options:
     1. Bubble sort by ID
     2. Insertion sort by name
     3. Selection sort by CGPA (descending order)
     4. Exit program

3. **Sorting Algorithms**:
   - **Bubble Sort**: Used for sorting by student ID in ascending order
   - **Insertion Sort**: Used for sorting by student name in alphabetical order
   - **Selection Sort**: Used for sorting by CGPA in descending order (highest first)

4. **Display Function**:
   - A unified display function that shows student data in a formatted table with columns for ID, Name, Department, and CGPA
   - Shows both unsorted and sorted versions of the data for comparison

## Program Flow

1. Initializes with a predefined list of 10 student records
2. Enters an infinite loop displaying the menu options
3. Based on user input:
   - Performs the selected sorting operation
   - Displays the results in a tabular format
   - For invalid inputs, shows an error message
4. Continues until the user selects the exit option (4)

## Technical Details

- **Bubble Sort Implementation**: Compares adjacent elements and swaps them if they are in the wrong order, repeating until the list is sorted
- **Insertion Sort Implementation**: Builds the final sorted array one item at a time by inserting each new item into its proper position
- **Selection Sort Implementation**: Repeatedly finds the maximum element (for descending order) and moves it to the sorted portion
- **Display Formatting**: Uses tab characters (`\t`) for column alignment in the output

The program demonstrates fundamental sorting algorithms applied to real-world data management, with clear separation of concerns between sorting logic and display functionality.

```python

students = [
    {"id": 'DBU1601452', "name": "Petros ", "dept": "software", "cgpa": 4.0},
    {"id": 'DBU1601603', "name": "Behailu", "dept": "computer science", "cgpa": 3.96},
    {"id": 'DBU1601333', "name": "Abebe", "dept": "engineering", "cgpa": 2.8},
    {"id": 'DBU1601201', "name": "Debebe", "dept": "Medicine", "cgpa": 3.9},
    {"id": 'DBU1601589', "name": "Eyerus", "dept": "Civil enginering", "cgpa": 3.2},
    {"id": 'DBU1601108', "name": "Fikrte", "dept": "computer science", "cgpa": 3.7},
    {"id": 'DBU1601093', "name": "Gebrye", "dept": "information.te", "cgpa": 2.9},
    {"id": 'DBU1601687', "name": "Helina", "dept": "computer science", "cgpa": 3.3},
    {"id": 'DBU1601060', "name": "Gemechu", "dept": "Data science", "cgpa": 3.6},
    {"id": 'DBU1601471', "name": "goitom", "dept": "information.te", "cgpa": 3.1}
]
while True:
    print("@@@@@@@@@@@@@@@@@@@@@\nplease choice sorting method !!!\n1: Bubble sort(by ID)\n2: insertion sort(by name)\n3: selection sort(by CGPA)\n4: exit\n@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
    ch=int(input("what you want: "))
    if ch==1:
        def sort_students_by_id(ar):
            for i in range(len(ar)-1):
                for j in range(0, len(ar)-i-1):
                    if ar[j]["id"] > ar[j+1]["id"]:
                        ar[j], ar[j+1] = ar[j+1], ar[j]

        # Display function with headers
        def display_students(title, student_list):
            print(f"\n{title}")
            print("ID\t\tName\t\tDepartment\t\t\tCGPA")
            for s in student_list:
                print(f"{s['id']}\t{s['name']}\t\t{s['dept']}\t\t\t{s['cgpa']}")

        # Show unsorted
        display_students("Unsorted Students:", students)

        # Sort and show sorted
        sort_students_by_id(students)
        display_students("Sorted Students by ID:", students)


    elif ch==2:
        def sort_students_by_name(ar):
            for i in range(1, len(ar)):
                key = ar[i]
                j = i - 1
                while j >= 0 and ar[j]["name"] > key["name"]:
                    ar[j + 1] = ar[j]
                    j = j - 1
                ar[j + 1] = key

        # Display function with headers
        def display_students(title, student_list):
            print(f"\n{title}")
            print("ID\t\tName\t\tDepartment\t\t\tCGPA")
            for s in student_list:
                print(f"{s['id']}\t{s['name']}\t\t{s['dept']}\t\t\t{s['cgpa']}")

        # Show unsorted
        display_students("Unsorted Students:", students)

        # Sort and show sorted
        sort_students_by_name(students)
        display_students("Sorted Students by NAME:", students)

    elif ch==3:
        def sort_students_by_cgpa_desc(ar):
            for i in range(len(ar)):
                maxindex = i  
                for j in range(i + 1, len(ar)):
                    if ar[j]["cgpa"] > ar[maxindex]["cgpa"]:  # > for descending
                        maxindex = j
                if maxindex != i:
                    ar[i], ar[maxindex] = ar[maxindex], ar[i]
        # Display function with headers
        def display_students(title, student_list):
            print(f"\n{title}")
            print("ID\t\tName\t\tDepartment\t\t\tCGPA")
            for s in student_list:
                print(f"{s['id']}\t{s['name']}\t\t{s['dept']}\t\t\t{s['cgpa']}")

        # Show unsorted
        display_students("Unsorted Students:", students)

        # Sort and show sorted
        sort_students_by_cgpa_desc(students)
        display_students("Sorted Students by CGPA:", students)
       
    elif ch==4:
       print("thank you!!! the program is terminated !!!")
       exit()
    else:
        print("invalid input!!! please enter correct choice ???")
```

# Sorting Algorithms Performance Comparison

**Bubble Sort (by ID):**
- Works by repeatedly swapping adjacent elements
- Best for nearly-sorted data (O(n) in best case)
- Inefficient for random data (O(n²) average/worst case)
- Many unnecessary swaps make it slowest for most cases

**Insertion Sort (by Name):**
- Builds sorted portion one element at a time
- Excels with small or partially-sorted data (O(n) best case)
- Fewer comparisons than Bubble Sort in practice
- Fastest choice for this small student dataset

**Selection Sort (by CGPA):**
- Always takes O(n²) time regardless of input
- Minimizes swaps to exactly n-1 (most swap-efficient)
- Performance remains consistent for all cases
- Simple but never outperforms others for small data

All algorithms:
- Use constant O(1) memory (in-place sorting)
- Show negligible speed differences for 10 records
- Would be too slow for large datasets (n > 1000)
- Primarily used here for educational demonstration

For real-world applications with larger data, merge sort or Python's built-in sorted() (Timsort) would be better choices. These O(n log n) algorithms scale much better than the quadratic ones implemented here.

# advantage and disadvantages

### **1. Bubble Sort (by ID)**
**Advantages:**
- Simple to implement and understand
- Works well for nearly-sorted data (e.g., if IDs are mostly in order)
- Stable (maintains relative order of equal keys)
- Performs in-place sorting (O(1) extra memory)

**Disadvantages:**
- Very slow for large datasets (O(n²) time)
- Many unnecessary swaps even when data is already sorted
- Poor performance on reverse-sorted data
- Not suitable for real-world applications with large dictionaries

---

### **2. Insertion Sort (by Name)**
**Advantages:**
- Efficient for small datasets (ideal for <50 elements)
- Adaptive (faster when data is partially sorted)
- Stable sorting (preserves order of equal names)
- Low overhead and simple implementation
- Performs well when adding new data to an already sorted list

**Disadvantages:**
- Still O(n²) in worst case (slow for large datasets)
- Requires shifting elements, which can be costly
- Bad for reverse-sorted data (worst performance)

---

### **3. Selection Sort (by CGPA)**
**Advantages:**
- Minimizes swaps (only O(n) swaps, good if writes are expensive)
- Simple to implement
- Consistent performance (always O(n²), regardless of input order)
- In-place sorting (no extra memory needed)

**Disadvantages:**
- Always slow (O(n²) even for sorted data)
- Not stable (may change order of equal CGPAs)
- Poor performance compared to Insertion Sort for small datasets
- No early termination (always checks all elements)

---

### **Best Choice for Dictionary Data?**
- **For small datasets (≤50 items):** **Insertion Sort** (fast and stable)  
- **For nearly-sorted data:** **Bubble Sort** (early termination helps)  
- **For minimizing swaps:** **Selection Sort** (but rarely the best choice)  
- **For real-world use:** Python’s built-in `sorted()` (Timsort, hybrid algorithm) is far superior for large dictionaries.  

**Conclusion:**  
For the given student records (10 entries), **Insertion Sort** is the most efficient, while **Bubble Sort** and **Selection Sort** serve mainly for educational purposes. None are ideal for large-scale data sorting.
